# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
'''sch.linkage(X, method='ward'): This computes the hierarchical clustering using the Ward's method and returns the linkage matrix.
sch.dendrogram(): This function plots the dendrogram using the linkage matrix.'''
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

## Visualising the clusters

```python
# Visualising the clusters
'''X is your dataset containing the features (Annual Income and Spending Score).
y_hc contains the cluster labels obtained from a clustering algorithm (like Hierarchical Clustering).
X[y_hc == i, 0] selects the Annual Income values for the i-th cluster.
X[y_hc == i, 1] selects the Spending Score values for the i-th cluster.
s = 100 sets the size of the scatter plot points.
c sets the color of the points for each cluster.
label provides a label for each cluster which will be shown in the legend.'''
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Carefull')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'General')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Target')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Careless')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Miser')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
